```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv('results_get_popular.csv')
```

```python
# plot the results
threads = data['threads']
throughtput = data['throughput']
plt.plot(threads, throughtput)
plt.xlabel('Number of Concurrent Users')
plt.ylabel('Throughput (requests/second)')
plt.title('Throughput vs Number of Concurrent Users')
plt.grid(True)
plt.xlim(0, max(threads) + 1)
plt.ylim(0, max(throughtput)+100)
```

```python
# plot the results
threads = data['threads']
response_time = data['response_time']
plt.plot(threads, response_time)
plt.xlabel('Number of Concurrent Users')
plt.ylabel('Response Time (ms)')
plt.title('Response Time vs Number of Concurrent Users')
plt.grid(True)
plt.ylim(0, max(response_time))
plt.xlim(0, max(threads) +1)
```

```python
# plot the results
threads = data['threads']
cpu_usage = data['cpu_util']
plt.plot(threads, cpu_usage)    
plt.xlabel('Number of Concurrent Users')
plt.ylabel('CPU Utilization (%)')
plt.title('CPU Utilization vs Number of Concurrent Users')
plt.grid(True)
plt.ylim(0, 100)
plt.xlim(0, max(threads) + 1)
```

```python
# plot the results
threads = data['threads']
disk_util = data['disk_util']
plt.plot(threads, disk_util)
plt.xlabel('Number of Concurrent Users')
plt.ylabel('Disk Utilization (%)')
plt.title('Disk Utilization vs Number of Concurrent Users')
plt.grid(True)
plt.ylim(0, 5)
plt.xlim(0, max(threads) + 1)
```
